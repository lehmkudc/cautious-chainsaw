```python
os.listdir("data")
output_data = pd.DataFrame()

for file in os.listdir("data/"):
    if file.endswith(".csv"):
        subset = pd.read_csv("data/" + file)
        output_data = output_data.append( subset )
    
print( output_data.shape )
output_data.head()
```
