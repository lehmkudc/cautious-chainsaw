```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import simps, trapz, cumtrapz
from scipy.ndimage.filters import gaussian_filter1d
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 9]
import os
```

```python
os.listdir("data")
output_data = pd.DataFrame()

for file in os.listdir("data/"):
    if file.endswith(".csv"):
        subset = pd.read_csv("data/" + file)
        output_data = output_data.append( subset )
    
print( output_data.shape )
output_data.head()
```

```python
ys = [0.0, 0.25, 0.5, 0.75, 1.0]
Ts = [30.0, 45.0, 60.0]
target_data = output_data
```

```python
for y in ys:
    target_data = output_data[output_data.yco_res == y]
    plt.scatter(
        x = target_data.P_res,
        y = target_data.P_sim,
        label = "yco:" + str(y)
    )
plt.legend()
```

```python
for T in Ts:
    target_data = output_data[output_data.T_res == T]
    plt.scatter(
        x = target_data.P_res,
        y = target_data.P_sim,
        label = "T:" + str(T)
    )
plt.legend()
```

```python
for T in Ts:
    target_data = output_data[output_data.T_res == T]
    target_data = target_data[target_data.yco_res == 1.0]
    plt.scatter(
        x = target_data.P_res,
        y = target_data.P_sim,
        label = "T:" + str(T)
    )
plt.legend()
```
